```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
data = pd.read_csv('STA160_Midterm_Data_Processed.csv', dtype={'Response': str})
```

```python
# Empirical CDFs for each variable
```

```python
# Binary Variables
```

```python
# Find Empirical CDF for High BP
res = stats.ecdf(data['HighBP'])
ax = plt.subplot()
res.cdf.plot(ax)
ax.set_xlabel('HighBP')
ax.set_ylabel('Empirical CDF')
plt.show()
```
